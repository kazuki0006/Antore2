```python
# ä»Šæ—¥ã®æ—¥ä»˜ã‚’å–å¾—ã—ã€æ–‡å­—åˆ—ã¨ã—ã¦è¡¨ç¤ºï¼ˆYYYY-MM-DDå½¢å¼ï¼‰
today = datetime.date.today()
today_str = today.strftime('%Y/%m/%d')

# ãƒ‡ãƒ¼ã‚¿ã‚’æ ¼ç´ã™ã‚‹ãŸã‚ã®ãƒªã‚¹ãƒˆã‚’ä½œæˆ
data_work = []

# ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°ã®å®šç¾©: å‹¤å‹™åœ°ã¾ã§ã®æ™‚é–“
def calculation_work(duration_work): # ğŸ‘ˆç›´ç·šçš„ã§ãªãæ›²ç·šçš„ãªå‹¾é…
    if duration_work <= 10:
        return 100
    elif duration_work <= 15:
        return 90
    elif duration_work <= 20:
        return 80
    elif duration_work <= 30:
        return 70
    elif duration_work <= 40:
        return 60
    elif duration_work <= 50:
        return 50
    elif duration_work <= 60:
        return 40
    else:
        return 30

# å„ä½æ‰€ã«å¯¾ã—ã¦å‡¦ç†ã‚’ç¹°ã‚Šè¿”ã™
for address in addresses:
    nav_request = 'language=ja&origin={}&destination={}&key={}'.format(address, workplace, api_key) # ğŸ‘ˆé›»è»Šã§ã®å‡ºåŠ›ã¯ä¸€æ„ã«ãªã‚‰ãšDFåŒ–ãŒé›£ã—ã„ â‡’ ä¿®æ­£ã‚’æ¤œè¨ (å„ªå…ˆåº¦ä½)
    nav_request = urllib.parse.quote_plus(nav_request, safe='=&')
    request = endpoint + nav_request

    # Google Maps Platform Directions APIã‚’å®Ÿè¡Œ
    response = urllib.request.urlopen(request).read()

    # çµæœ(JSON)ã‚’å–å¾—
    directions = json.loads(response)

    # æ‰€è¦æ™‚é–“ã‚’å–å¾—
    for key in directions['routes']:
        for key2 in key['legs']:
            distance_work = key2['distance']['text']
            duration_work = key2['duration']['text']

            # è·é›¢ã®æ–‡å­—åˆ—ã‹ã‚‰"km"ã‚’å–ã‚Šé™¤ã„ã¦æ•°å€¤ã«å¤‰æ›
            distance_work = float(distance_work.replace(' km', '').replace(',', ''))

            # æ™‚é–“ã®æ–‡å­—åˆ—ã‹ã‚‰"åˆ†"ã‚’å–ã‚Šé™¤ã„ã¦æ•°å€¤ã«å¤‰æ›
            duration_work = int(duration_work.replace('åˆ†', '')) # ğŸ‘ˆ1æ™‚é–“ã‚’è¶…ãˆã‚‹ã¨"xxæ™‚é–“"ã¨è¡¨ç¤ºã•ã‚Œã‚‹ãŸã‚ã‚¨ãƒ©ãƒ¼ã«ãªã‚‹ã€‚è¦ä¿®æ­£

            # ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°: å‹¤å‹™åœ°ã¾ã§ã®æ™‚é–“
            score_work = calculation_work(duration_work)
            
            # å¿…è¦ãªãƒ‡ãƒ¼ã‚¿ã‚’ãƒªã‚¹ãƒˆã«æ ¼ç´
            data_work.append([address, score_work, distance_work, duration_work])

# ãƒ‡ãƒ¼ã‚¿ã‚’DataFrameã«å¤‰æ›
df_work = pd.DataFrame(data_work, columns=['å€™è£œç‰©ä»¶ä½æ‰€', 'åƒãã‚„ã™ã•ã‚¹ã‚³ã‚¢', 'ç›®çš„åœ°ã¾ã§ã®è·é›¢(km)', 'åˆ°ç€ã¾ã§ã«ã‹ã‹ã‚‹æ™‚é–“(åˆ†)'])

# ãƒ‡ãƒ¼ã‚¿ã‚’ã‚¢ã‚¦ãƒˆãƒ—ãƒƒãƒˆ
df_work
```

2. ã‚¢ã‚¦ãƒˆãƒ—ãƒƒãƒˆ - â‘¡ æš®ã‚‰ã—ã‚„ã™ã• - A ã‚¹ãƒ¼ãƒ‘ãƒ¼

```python
### A: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã®ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚° ###

## è¨­å®šãƒ»æº–å‚™ ##

# ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã®ç”Ÿæˆ
gmaps = googlemaps.Client(key=api_key)

# æ–½è¨­åã®æŒ‡å®š
place_name = "ã‚¹ãƒ¼ãƒ‘ãƒ¼" # ğŸ‘ˆæ–½è¨­åã‚’ç¢ºèª

# ãƒ‡ãƒ¼ã‚¿ã‚’æ ¼ç´ã™ã‚‹ãŸã‚ã®ãƒªã‚¹ãƒˆã‚’ä½œæˆ
data_life_supermarket = []

# ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°ã®å®šç¾©: æ–½è¨­æ•°
def calculation_life_supermarket_number(number_supermarket): # ğŸ‘ˆãƒ­ã‚¸ãƒƒã‚¯ã‚’ç¢ºèª
    if number_supermarket >= 3:
        return 100
    elif 2 <= number_supermarket < 3:
        return 75
    elif 1 <= number_supermarket < 2:
        return 50
    else:
        return 30

# ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°ã®å®šç¾©: æ‰€è¦æ™‚é–“
def calculation_life_supermarket_duration(nearest_supermarket_duration): # ğŸ‘ˆãƒ­ã‚¸ãƒƒã‚¯ã‚’ç¢ºèª
    duration_int = int(nearest_supermarket_duration.replace("åˆ†", ""))
    return 100 - (duration_int - 1)*5

# æ¤œç´¢ç¯„å›²ã®æŒ‡å®š
radius = 800 # ğŸ‘ˆæ–½è¨­ã«å¿œã˜ã¦èª¿æ•´


## ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚° ##

# å„ä½æ‰€ã«å¯¾ã—ã¦å‡¦ç†ã‚’ç¹°ã‚Šè¿”ã™
for address in addresses:
    # ä½æ‰€ã‹ã‚‰ç·¯åº¦çµŒåº¦ã‚’å–å¾—
    geocode_result = gmaps.geocode(address)
    location = geocode_result[0]['geometry']['location']

    # æ¤œç´¢çµæœã‚’å–å¾—
    places_result = gmaps.places_nearby(location=location, radius=radius, keyword=place_name, language='ja')

    # æ¤œç´¢ç¯„å›²ã®æ–½è¨­æ•°ã‚’æ ¼ç´ã™ã‚‹å¤‰æ•°ã‚’ä½œæˆ
    number_supermarket = 0

    # æ¤œç´¢çµæœã‚’ãƒ«ãƒ¼ãƒ—ã—ã¦æ¤œç´¢ç¯„å›²ã®æ–½è¨­æ•°ã‚’ã‚«ã‚¦ãƒ³ãƒˆ
    for place in places_result['results']:
        # æ–½è¨­ã¾ã§ã®è·é›¢(å¾’æ­©)ã‚’å–å¾—
        distance_matrix_result = gmaps.distance_matrix(location, place['geometry']['location'], mode='walking', language='ja', units='metric')
        distance_supermarket = distance_matrix_result['rows'][0]['elements'][0]['distance']['value']

        # æ¤œç´¢ç¯„å›²ã®å ´åˆã«ã‚«ã‚¦ãƒ³ãƒˆ
        if distance_supermarket <= radius:
            number_supermarket += 1

    # ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°: æ–½è¨­æ•°
    score_life_supermarket_number = calculation_life_supermarket_number(number_supermarket)
        
    # æœ€ã‚‚è¿‘ã„æ–½è¨­ã¾ã§ã®æ‰€è¦æ™‚é–“ã‚’æ•´æ•°ã«å¤‰æ›ã—ã¦ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°
    nearest_supermarket_distance = float('inf')
    nearest_supermarket_duration = None
    nearest_supermarket_name = None
    
    for place in places_result['results']:
        # æ–½è¨­ã¾ã§ã®æ‰€è¦æ™‚é–“ã‚’å–å¾—
        distance_matrix_result = gmaps.distance_matrix(location, place['geometry']['location'], mode='walking', language='ja', units='metric')
        duration_supermarket = distance_matrix_result['rows'][0]['elements'][0]['duration']['text']

        # æœ€ã‚‚è¿‘ã„æ–½è¨­ã‚’æ›´æ–°
        if 'distance' in distance_matrix_result['rows'][0]['elements'][0]:
            distance_supermarket = distance_matrix_result['rows'][0]['elements'][0]['distance']['value']
            if distance_supermarket < nearest_supermarket_distance:
                nearest_supermarket_distance = distance_supermarket
                nearest_supermarket_duration = duration_supermarket
                nearest_supermarket_name = place['name']
    
    # ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°: æ‰€è¦æ™‚é–“
    score_life_supermarket_duration = calculation_life_supermarket_duration(nearest_supermarket_duration)
                
    #ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°: æ–½è¨­æ•°ï¼‹æ‰€è¦æ™‚é–“
    score_life_supermarket = (score_life_supermarket_number + score_life_supermarket_duration) / 2
            
    # å¿…è¦ãªãƒ‡ãƒ¼ã‚¿ã‚’ãƒªã‚¹ãƒˆã«æ ¼ç´
    data_life_supermarket.append([address, score_life_supermarket, score_life_supermarket_number, score_life_supermarket_duration, number_supermarket, nearest_supermarket_name, nearest_supermarket_duration])

# ãƒ‡ãƒ¼ã‚¿ã‚’DataFrameã«å¤‰æ›
df_life_supermarket = pd.DataFrame(data_life_supermarket, columns=['å€™è£œç‰©ä»¶ä½æ‰€', 'ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¹ã‚³ã‚¢', 'ã‚¹ãƒ¼ãƒ‘ãƒ¼æ•°ã‚¹ã‚³ã‚¢', 'ã‚¹ãƒ¼ãƒ‘ãƒ¼æ‰€è¦æ™‚é–“ã‚¹ã‚³ã‚¢', 'ã‚¨ãƒªã‚¢å†…ã®ã‚¹ãƒ¼ãƒ‘ãƒ¼æ•°', 'æœ€ã‚‚è¿‘ã„ã‚¹ãƒ¼ãƒ‘ãƒ¼å', 'æœ€ã‚‚è¿‘ã„ã‚¹ãƒ¼ãƒ‘ãƒ¼ã¾ã§ã®æ‰€è¦æ™‚é–“(åˆ†)'])
    
# ãƒ‡ãƒ¼ã‚¿ã‚’ã‚¢ã‚¦ãƒˆãƒ—ãƒƒãƒˆ
df_life_supermarket
```
